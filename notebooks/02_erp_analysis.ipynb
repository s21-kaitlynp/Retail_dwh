```python
import pandas as pd

print("Анализируем ERP данные...")
print("Файлы в data/:")
import os
print(os.listdir("data"))
```

```python
erp_files = [f for f in os.listdir("data") if f.isupper() and f.endswith('.csv')]
print(f"ERP файлы: {erp_files}")

for file in erp_files:
    df = pd.read_csv(f"data/{file}")
    print(f"\n=== {file} ===")
    print(f"Размер: {df.shape}")
    print("Колонки:", list(df.columns))
    print("\nПервые 3 строки:")
    print(df.head(3))
```
